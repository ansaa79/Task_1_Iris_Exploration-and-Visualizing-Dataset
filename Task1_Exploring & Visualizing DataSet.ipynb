# Task 1: Exploring and Visualizing the Iris Dataset

## Problem Statement
The goal of this task is to load, inspect, and visualize the Iris dataset using Python. 
This includes checking data quality, generating summary statistics, and creating plots to understand feature relationships.

## Objective
- Learn to load and inspect a dataset using pandas.
- Visualize data distributions and relationships using matplotlib and seaborn.
- Identify trends, patterns, and outliers in the dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Show plots inside the notebook
%matplotlib inline

# set a nice style for plots
sns.set_theme(style="whitegrid")
```

```python
# Load iris dataset CSV
df = pd.read_csv("iris.csv")

# First few rows
df.head()
```

```python
# Print shape of dataset (rows, columns)
print("Dataset shape:", df.shape)

# Print column names
print("Column names:", df.columns.tolist())
```

```python
# Summary info of dataset
print("Dataset info:")
df.info()

# Descriptive statistics
print("\n\nSummary statistics:")
df.describe()
```

```python
#Scatter Plots
plt.figure(figsize=(6,4))
sns.scatterplot(x="SepalLengthCm", y="SepalWidthCm", hue="Species", data=df)
plt.title("Sepal Length vs Sepal Width")
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(x="PetalLengthCm", y="PetalWidthCm", hue="Species", data=df)
plt.title("\n\nPetal Length vs Petal Width")
plt.show()
```

```python
# Histogram of petal length
plt.figure(figsize=(6,4))
sns.histplot(df["PetalLengthCm"], bins=20, kde=True)
plt.title("Distribution of Petal Length")
plt.show()

# Histogram for all numeric features
df.hist(figsize=(10, 8), bins=10)  # Increase figure size and set number of bins
plt.suptitle("Histograms of All Features", fontsize=16)  # Add title
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust spacing to avoid overlap with title
plt.show()
```

```python
#Box plots
plt.figure(figsize=(6,4))
sns.boxplot(x="Species", y="PetalWidthCm", data=df)
plt.title("Box Plot of Petal Width by Species")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x="Species", y="SepalLengthCm", data=df)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```

## Explanation of Results
- **Data quality:** No missing values detected; dataset is clean.
- **Scatter plots:** Petal length and width clearly separate species. Sepal measurements overlap more.
- **Histograms:** Show distribution of numeric features and help understand ranges and central tendency.
- **Box plots:** Identify outliers and feature differences between species.
